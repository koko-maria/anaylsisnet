```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Netflix = pd.read_csv("netflix_titles.csv")
Netflix.head()
```

```python
Netflix.columns
```

```python
Netflix.info
```

#### Drop "show_id" column 

```python
Netflix = Netflix.drop('show_id', axis = 1)
```

```python
Netflix.head()
```

```python
Netflix.info()
```

```python
Netflix['type'].value_counts()
```

```python
Netflix.isnull().sum()
```

```python
Netflix = Netflix.dropna(subset = ['duration','date_added','rating'], axis = 0)
```

```python
Netflix.info()
```

```python
Netflix['director'] = Netflix['director'].fillna('Not Specified')
Netflix['cast'] = Netflix['cast'].fillna('Not Specified')
Netflix['country'] = Netflix['country'].fillna('Not Specified')
Netflix.head(10)
```

```python
Netflix['type'].value_counts()
```

```python
#A pie chart showing the distribution of netflix movies and netflix tv shows.



plt.pie(Netflix['type'].value_counts(),
        labels=Netflix['type'].value_counts().index,
        colors=['orange', 'Blue'], startangle=90)
plt.legend();
```

### Movie subset

```python
Netflix_movie = Netflix[Netflix['type'] == 'Movie']
Netflix_movie.head(5)
```

```python
Netflix_movie['country'].value_counts()
```

```python
# Split the movie directors to only include 1 name
movie_country = Netflix_movie['country'].str.split(',', expand=True).stack()
# convert to a dataframe
movie_country = pd.DataFrame(movie_country)
movie_country.columns = ['Country']
movie_country.head() 
```

```python
# only selecting the known directors
movie_country = movie_country[movie_country['Country'] != 'Not Specified']

movie_country = movie_country.groupby(['Country']).size().reset_index(name='Total Movies')
```

```python
top_Countries = pd.DataFrame(movie_country.sort_values(by=['Total Movies'], ascending=False)).head(10)
top_Countries
```

```python
plt.figure(figsize=(14, 10))
sns.barplot(y='Country', x='Total Movies', data=top_Countries, color='turquoise')
plt.title('Top movie countries');
```

```python

movie_release_year = Netflix_movie['release_year']
# convert to a dataframe
movie_release_year = pd.DataFrame(movie_release_year)
movie_release_year.columns = ['Year Released']
movie_release_year.head(19) 
```

```python

movie_release_year = movie_release_year.groupby(['Year Released']).size().reset_index(name='Total Movies')
```

```python
highest_release_year = pd.DataFrame(movie_release_year.sort_values(by=['Total Movies'], ascending=False)).head(10)
highest_release_year
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(y='Total Movies', x='Year Released', data=highest_release_year, color='turquoise')
plt.title('Highest Movie Release_year');
```

```python
# Split the movie directors to only include 1 name
movie_director = Netflix_movie['director'].str.split(',', expand=True).stack()
# convert to a dataframe
movie_director = pd.DataFrame(movie_director)
movie_director.columns = ['director']
movie_director.head() 
```

```python
# only selecting the known directors
movie_director = movie_director[movie_director['director'] != 'Not Specified']

movie_director = movie_director.groupby(['director']).size().reset_index(name='Total Movies')
```

```python
top_director = pd.DataFrame(movie_director.sort_values(by=['Total Movies'], ascending=False)).head(10)
top_director
```

```python
plt.figure(figsize=(14, 10))
sns.barplot(y='director', x='Total Movies', data=top_director, color='blue')
plt.title('Top movie director');
```
